## 1. Gauss's law and cavities

2. charge density

Total charge $+q$ is spread across the surface. Sphere surface area $= 4\pi r^2$


Sphere: $\sigma_R = \dfrac{+q}{4\pi R^2}$

Inside shell: $\sigma_a = \dfrac{-q}{4\pi a^2}$ <br>
Outside shell: $\sigma_b = \dfrac{+q}{4\pi b^2}$

3. Electric field

$E$ inside conductor is always zero for our purposes. Use gauss's law to find E-fields outside conductors.

$$
\int_{\mathcal{S}} E \cdot da = Q_{enc}/\epsilon_0\\
E(r) = \begin{cases}
0 & r\leq R \\
\dfrac{+q}{4\pi r^2 \epsilon_0}  & R < r < a \\
0 & a \leq r\leq b \\
\dfrac{+q}{4\pi r^2 \epsilon_0}  & r > b \\

\end{cases}
$$

Sketch:



4. Voltage

$$
\Delta V = -\int E \cdot ds
$$

The integral is multipart

$$
\int_\infty^{r>b} E_r dr = \frac{q}{4\pi \epsilon_0} \left. \left(-\frac{1}{r}\right) \right|_\infty^r

=-\frac{q}{4\pi \epsilon_0 r} \\

\int_b^a E_r dr = 0 \\

\int_a^{r>R} E_r dr = \frac{q}{4\pi \epsilon_0} \left. \left(-\frac{1}{r}\right) \right|_a^r \\

=- \frac{q}{4\pi \epsilon_0 r} + \frac{q}{4\pi \epsilon_0 a} \\

\int_R^0 E_r dr = 0
$$

We have our $\Delta V$ for each case, so we can assemble them together with $V(\infty)=0$ to get our voltage function

$$
V(r) = \begin{cases}

\frac{q}{4\pi \epsilon_0} \left(-\frac{1}{R} + \frac{1}{a}
-\frac{1}{b} \right) & r\leq R\\
\frac{q}{4\pi \epsilon_0} \left(-\frac{1}{r} + \frac{1}{a}
-\frac{1}{b} \right) & R<r<a\\
-\frac{q}{4\pi \epsilon_0 b} & a\leq r \leq b \\
-\frac{q}{4\pi \epsilon_0 r} & r>b\\
\end{cases}
$$

Notice that $V(r)$ is continuous, and at $r=a$ the potential simplifies down to $-\frac{q}{4\pi \epsilon_0 b}$. Continuity is not disrupted by the charge densities, but the voltage function is not smooth

## 2. Coax capacitors





1. find $\Delta V$

Our result will be length-indepenedent. We can use the Poisson equation, but I'll just go through electric field first for old time's sake.

Using cylinder gaussian shells of length $L$ and radius $a<r<b$, we have

$$
\int E \cdot da = \frac{Q_{enc}}{\epsilon_0} \\
E_r 2\pi r l = \frac{\lambda l}{\epsilon_0} \quad \\
E(r) = \frac{\lambda}{2 \pi r \epsilon_0}
$$

The boundaries $a$ and $b$ are used since this is where $Q_{enc} = \lambda l$. For $r<a$ there is no charge (it rests on the surface of the conductor), and for $r>b$ the charge enclosed from the two conductors cancels since they have the opposite magnitude linear density. Next is voltage

$$
\Delta V = V(b) - V(a) = -\int_a^b E(r) dr  \\
= -\frac{\lambda}{2\pi \epsilon_0} \int_a^b \frac{1}{r} dr \\
= -\frac{\lambda}{2\pi \epsilon_0} \ln\left(\frac{a}{b} \right) \\
= \frac{\lambda}{2\pi \epsilon_0} \ln\left(\frac{b}{a} \right)
$$

2. Energy method 1

Since $a<b$, the logarithm of $a/b$ is negative and $b/a$ is positive. The charge separated between the two conductors on segments of length $l$ are $\lambda l$, so capacitance is

$$
C = \frac{\lambda l}{\Delta V} = \frac{\lambda l * 2\pi \epsilon_0}{\lambda \ln\left(\frac{b}{a} \right)} = \frac{2\pi l \epsilon_0}{\ln\left(\frac{b}{a} \right)}
$$
(Note: $C/l$ just remove $l$)

Capacitance should be dependent on the geometry of a capacitor, not total charge on the surface, so we're happy that $\lambda$ falls out here. Last is energy

$$
W = \frac{1}{2} C \left( \Delta V \right)^2 
= \frac{\pi l \epsilon_0}{\ln\left(\frac{b}{a} \right)} \left(\frac{\lambda}{2\pi \epsilon_0} \ln\left(\frac{b}{a} \right)\right)^2 
= l\frac{\lambda^2 \ln\left(\frac{b}{a}\right)}{4\pi \epsilon_0} \\
\frac{W}{l} = \frac{\lambda^2 \ln\left(\frac{b}{a}\right)}{4\pi \epsilon_0}
$$

Proof.

We will show that for two solutions to the same BVP, $V_1$ and $V_2$, that the difference between the two $V_3 = V_1 - V_2 = 0$.

Since each potential depends on the same charge distribution and boundary conditions, we have

$$
\nabla^2 V_1 = \nabla^2 V_2 = -\frac{\rho}{\epsilon_0}
$$

so the laplacian of $V_3$ is 

$$
\nabla^2 V_3 = \nabla^2\left(V_1 - V_2 \right) = -\frac{\rho}{\epsilon_0} + \frac{\rho}{\epsilon_0} = 0
$$

According to Green's identity, we can find the relationship between the volume integral of two arbitrary scalar potentials and the surface integral of the two as

$$
\int_V \left( T\nabla^2 U + \nabla T \cdot \nabla U\right)d\tau = 
\oint_S \left(  T\nabla U \right) \cdot dA
$$

Using $T=U=V_3$, this gives us

$$
\int_V \left(V_3 \nabla^2 V_3 + |\nabla V_3|^2 \right) d\tau = \oint_S\left( V_3 \nabla V_3\right) \cdot dA \\
\rightarrow \int_V |\nabla V_3|^2 d\tau = \oint_S\left( V_3 \nabla V_3\right) \cdot dA
$$

Since $V_1$ and $V_2$ fit the same boundary conditions, they are equivalent on any surface. As such 
$$
\oint_S\left( V_3 \nabla V_3\right) \cdot dA = 0 \\
\rightarrow \int_V |\nabla V_3|^2 d\tau = 0
$$

Since $|\nabla V_3|^2$ is always positive, $V_3=0$ is necessary for this integral to be true as any value $V_3 \neq 0$ would result in some contribution to the volume integral. Thus $V_1$ and $V_2$ are equivalent for the shared charge distribution. QED
