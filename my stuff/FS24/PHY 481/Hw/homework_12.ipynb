Symmetric about z-axis. Use an amperian loop

$$
\oint \vec B \cdot d\vec l = \mu_0 \int \vec J \cdot d\vec a
$$

$\vec B$ is zero up and down. Use recangular loop of height $z$ and arbitrary length $L$.

$$
\oint \vec B \cdot d\vec l = \int_0^L B(z) dL -\int_{L}^0  B(z)dl = 2LB(z) \\
$$
Next the area of the loop
$$
d\vec a = dy dz \hat x \\
\vec J(z) = J_0 \text{abs}(z)x \\
\int \vec J(z) \cdot d\vec a = \int_{-z}^z\int_{0}^L J_0 |z| dy dz = 2\int_0^z\int_{0}^L J_0 z dy dz \\
= 2J_0L \int_0^z z dz = \mu_0 J_0 L z^2 
$$
Combine them together
$$
2LB(z) = \mu_0 J_0 L z^2  \\ B(z) = \frac{\mu_0 J_0 z^2}{2} \\
\vec B(z) = \begin{cases} -B(z) \hat y & z\geq 0 \\ B(z) \hat y  & z < 0\end{cases}
$$
Vector signs/directions are determined via right-hand-rule

For the field outside the slab: area integral stops at $z=\pm h$ since $\vec J(|z|>h)=0$.

$$
\int \vec J(z) \cdot d\vec a = 2J_0L \int_0^h z dz = \mu_0 J_0 L h^2 \\
B = \frac{\mu_0 J_0 L h^2}{2} \\
\vec B(z) = \begin{cases} -\frac{\mu_0 J_0 L h^2}{2} \hat y & z\geq 0 \\ \frac{\mu_0 J_0 L h^2}{2} \hat y  & z < 0\end{cases}
$$
Notice there is no $z$ component outside. As we've seen with electric fields, the magnetic field extends forever for a flat infinite plane
