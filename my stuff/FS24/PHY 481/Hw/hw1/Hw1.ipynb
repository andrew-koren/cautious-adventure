The vector perpenicular to a plane is the cross product of two vectors on the surface of the plane. Given the verticies of the triangle in the figure, we can create vectors on the surface

$$
\vec a = (1, 0, 0) \rightarrow (0, 0, 3) \\
\vec b = (1, 0, 0) \rightarrow (0, 2, 0)
$$

The separation vector for each is thus

$$
\vec a = \langle-1, 0, 3\rangle \\
\vec b = \langle-1, 2, 0\rangle \\
$$

The cross product is

$$
\vec a \times \vec b = \left| \begin{matrix} \hat x & \hat y & \hat z \\ -1 & 0 & 3 \\ -1 & 2 & 0\end{matrix} \right| = -2 \hat z -3\hat y
$$

Well, that clearly points in the wrong direction. Remember right hand rule...

$$
\vec b \times \vec a = 2 \hat z + 3\hat y
$$

A unit vector has magnitude 1, so 

$$
\hat n = \frac{2 \hat z + 3\hat y}{\left|2 \hat z + 3\hat y\right|} = \frac{2 \hat z + 3\hat y}{\sqrt{13}}
$$





Part 1: problem 1.7

The source and field vectors are

$$
\vec r_1 = \langle 2, 8, 7 \rangle \\
\vec r_2 = \langle 4, 6, 8 \rangle
$$

The separation vector is simply end minus beginning, so 

$$
\mathbb{r} = \vec r_2 - \vec r_1 = \langle 2, -2, 1 \rangle
$$

The magnitude is 

$$
\sqrt{2^2+(-2)^2+1^2} = \sqrt{9} = 3
$$

Which allows us to easily create the unit vector

$$
\hat{\mathbb{r}} = \frac{\mathbb{r}}{|\mathbb{r}|} = \left< \frac{2}{3}, \frac{-2}{3}, \frac{1}{3} \right>
$$

The divergence and curl operations are applicable to flowing fluid. At any point divergence describes if nearby vectors are net moving towards/away from that point, and could suggest that the fluid is becoming more/less dense or flowing into/out of the system near a given point. It is a scalar value with positive for vectors mostly moving away from the point, and negative if they are mostly moving toards. Curl describes whether fluid is flowing around a given point, and yields the axis which it is rotating around. Depending on perspective, a curl vector typically points in the direction which the fluid is rotating clockwise around, and a larger magnitude represents faster/greater flow.

$\mathbf{r} = (x-x')\hat x + (y-y')\hat y + (z-z') \hat z$

$$
|\mathbf{r}| = \sqrt{(x-x')^2+(y-y')^2+(z-z')^2} \\
\frac{\partial}{\partial x} |\mathbf{r}| = \frac{\partial}{\partial x} \sqrt{(x-x')^2+c}, \quad c = (y-y')^2+(z-z')^2
$$
Apply chain rule
$$
= \frac{1}{2} \left(\sqrt{(x-x')^2+c} \right)^{-1/2} \frac{\partial}{\partial x} (x - x')^2 +c \\
= \frac{x-x'}{\sqrt{(x-x')^2+c}} \\
$$

From this, the gradient is

$$
\nabla|\mathbf{r}| = \left(\hat x\frac{\partial}{\partial x} + \hat y\frac{\partial}{\partial y} +\hat z\frac{\partial}{\partial z} \right)|\mathbf{r}|
=
\hat x\frac{\partial}{\partial x}|\mathbf{r}| + \hat y\frac{\partial}{\partial y} |\mathbf{r}| +\hat z\frac{\partial}{\partial z} |\mathbf{r}| \\
= 
\frac{\hat x(x-x') + \hat y(y-y') + \hat z(z-z')}{\sqrt{(x-x')^2+(y-y')^2+(z-z')^2} }
