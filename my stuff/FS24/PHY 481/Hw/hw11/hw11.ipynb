## 1 - Current densities



Current should scale with radius. Note that charge density $\rho = Q/V = \frac{Q}{\frac{4}{3}\pi R^3}$



$$
\mathbf J = \rho \vec v \\
\vec v = \vec \omega \times \vec r \\
= r\omega \hat z \times \hat r \\
= r \omega \sin (\theta)  \hat \phi \\
\mathbf{J}(r,\theta,\phi)=\rho r \omega\sin (\theta)  \hat \phi = \frac{Q}{\frac{4}{3}\pi R^3}r \omega\sin (\theta)  \hat \phi
$$

Surface current density is quite easy, and we just work backwards from that to get volume current density

$$
\mathbf{K} = \sigma \vec v \\
\vec v = \vec \omega \times \vec s = s\omega \hat \phi \\
\mathbf{K}(s,\phi) = \sigma s\omega \hat \phi \\
\mathbf{J}(s,\phi,z) = \sigma s\omega \hat \phi \cdot\delta(z)
$$



This is an E&M question!

$$
\mathbf I(\phi) = \frac{Q}{2\pi R} \vec v = \frac{Q\omega}{2\pi} \hat \phi \\
$$
For $\mathbf{J}$, simply divide by the size of the ring and zero-in using $\delta$ functions
$$
L = 2\pi R \\
\mathbf{J}(s,\phi,z) = \frac{Q}{2\pi R} \omega R \hat \phi \cdot \delta(z)\delta(s-R)\\
=\frac{Q\omega}{2\pi} \hat \phi \cdot \delta(z)\delta(s-R)
$$
