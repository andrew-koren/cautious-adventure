```python
import numpy as np
import matplotlib.pyplot as plt

energy_in = np.linspace(0, 1, 1000)
eout_05 = energy_in*0.5
eout_1 = energy_in*1

plt.plot(energy_in, eout_05, label='R=0.5')
plt.plot(energy_in, eout_1, label='R=1')
plt.xlabel('Energy in')
plt.ylabel('Energy out')
plt.legend()
plt.title('Plot of R=0.5 and R=1')
```
