A simple cycle does not have repeating vertecies (aside from the start/finish).

The cycle $2437612$ is the longest, computed numerically using the networkx package below

```python
import networkx as nx
E = {(1,2),(1,6),(2,5),(2,7),(2,3),(2,4),(4,3),(3,7),(5,7),(6,7)}
G = nx.Graph(E)
nx.draw_networkx(G, with_labels = True)

print([cycle for cycle in nx.simple_cycles(G)])

a = sorted(list(nx.simple_cycles(G)), key = lambda s: -len(s)) 
print(a[0])
```

This result makes logical sense. To make a cycle that includes all points, either the node $7$ or $2$ must be hit twice since $61$, $34$ and $5$ are all connected only by $7$ and $2$.

example: see exercise 3.  

Codewords: $\{10, 11, 000, 001, 010, 011 \}$ <br>
Parameter: $(0, 0, 2, 4)$ <br>
Kraft-Mcmillan:
$$
K = 0 + \frac{0}{2} + \frac{2}{4} + \frac{4}{8} = 1
$$

This code is complete. Alphabet size $6$, and trivially divides evenly
$$
\frac{6-1}{2-1}=5
$$

We will proove that for a complete $b$-nary code, the encoded alphabet of size $m$ has the feature $m - 1 = 0 \pmod{b-1}$

b-nary code: $|T|=b$ <br>
alphabet of size m: $|S^n|=m$

Before continuing, note that this does not hold for $b=1$, and is trivial for $b=2$, since any value of $m-1$ is divisible by one. We will consider cases where $b>2$

Since $C: S^n \mapsto T^*$ exists and $K=1$,

$$
K = \sum_i^M\frac{n_i}{b^i} = 1 \\
$$
Where $M$ is the longest codeword in $C$, which exists since the alphabet being encoded has a finite size and $C$ is 1-to-1. <br> Then

$$
m = \sum_i^M n_i 
$$
and 
$$
b^M\sum_i^M\frac{n_i}{b^i} = b^M \\
\sum_i^M n_ib^{M-i} = b^M
$$




Since $b = 1 \pmod{b-1}$ and $b^k = 1 \pmod{b-1}$ for positive integers of $k$, this equates to

$$
\sum_i^M n_i = 1 \pmod{b-1} \\

m  = 1 \pmod{b-1} \\
m-1 = 0\pmod{b-1}
$$
Thus $m-1$ divides by $b-1$ with a remainder of zero

(b) The coefficients of $x^7$ represent how many codewords of $CC$  and $CCC$ (aka $C^3$) have length $7$. 

The list of messages in $S^*$ that map to codewords of length 16 and 6 are:

```python
code = {
    'a':'00',
    'b':'010',
    'c':'011',
    'd':'1000',
    'e':'1001',
    'f':'1101',
    'g':'1111'
}

q2 = set()
q3 = set()

for c1 in code.keys():
    for c2 in code.keys():
        if len(code[c1]+code[c2]) == 7:
            q2.add(c1+c2)

        for c3 in code.keys():
            if len(code[c1]+code[c2]+code[c3]) == 7:
                q3.add(c1+c2+c3)

print(q2)
print(q3)
 
```
