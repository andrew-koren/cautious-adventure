Electrons love to jiggle. We can model them in the tight-binding model.



Overlap -> tunneling.

Lattice >2 atoms

$$
H=\frac{\vec p^2}{2m} + \sum_n V(\vec r - \vec R'_n)
$$
Where $\vec R'_n$ is the position of the nucleus.

$|n\rangle$ is an eigenstate of a single nucleus, with each $n$ corresponding to an atom in the tight-binding model. 

To get to the eigenstates of the full hamiltonian, we need to consider the overlap

$$
\langle n | H | n \rangle = E_0 \\
\langle n | H | n \pm 1\rangle = -t \\
$$
Remember that $\langle n|n\pm 1\rangle $ is the overlap between the two. 

In our model, we can say $\langle n | H | m \rangle = 0$ where $|m-n| > 2$ <br>
More assumptions: translation invariance and orthogonality $\langle n | m \rangle = \delta_{nm}$

Notation:
$ H_{nm} \equiv \langle n | H |m \rangle = E_0 \delta_{nm} - t(\delta_{n,m+1} + \delta_{n,m-1})$

Let's diagonalize. Guess $\Psi_n \sim e^{-ikna}$

$$
\sum_m H_{nm} \Psi_m = E_n \Psi_n \\
\sum_m E_0 \delta_{nm} \Psi_m - t(\delta_{n,m+1}\Psi_m + \delta_{n,m-1}\Psi_m) = E_n \Psi_n\\
E_0 \Psi_n - t(\Psi_{n+1} + \Psi_{n-1}) = E_n\Psi_n
$$

Thanks to our guess, we can divide by $\Psi_n$

$$
E_0 - t\left(\Psi_1 + \Psi_{-1}\right) = E_n \\
E_0 - te^{-ika} - te^{ika} = E_n \\
E = E_0 -2t\cos(ka) \\
E = E(k)
$$

$E(k)$ dispersion relation

```python
import numpy as np
from numpy import cos, pi
import matplotlib.pyplot as plt

a = pi
E0 = 2
t = 1
params = [a, E0, t]

def E(k, params):
    a, E0, t = params
    return E0 - 2*t*cos(k*a)

k = np.linspace(-1.5, 1.5, 100)
plt.plot(k, E(k, params))
plt.annotate('Ground energy', [0,0], xycoords='data')
```

### Useful stuff

Non-degenerate perturbation theory corrects the energies $E_1, E_2, E_3$, etc by some small perturbation $H = H^0 + \lambda H$

Eigenstates and energies:
 $$
 H^{(0)}\left|k^{(0)}\right\rangle=E_k^{(0)}\left|k^{(0)}\right\rangle\\
 H(\lambda)|n\rangle_\lambda=E_n(\lambda)|n\rangle_\lambda
 $$
This yields
 $$
 E_n^{(1)} = V_{nn} =\langle n | V| n\rangle \\
 E_n^{(2)} = -\sum_{m\neq n} \frac{|V_{mn}|^2}{E_m^{(0)}-E_n^{(0)}}
 $$

The $|n\rangle^{(i)}$ subscripts represent the corrections, where all combined yield the actual perturbed eigenvalue $|n\rangle$

What if the energies are the same? Well we get a divide by zero <br>
In comes degenerate perturbation theory. We can separate energies which would otherwise be stacked on top of each other. <br> shitty notation inbound

 $$

A superposition of #1 and #2 yield the same energy. How do we determine the way they branch? Assuming $E_1^{(0)}=E_2^{(0)}=\dots$

Assume degeneracy is lifted by perturbation with the state $|k\rangle_\lambda$
