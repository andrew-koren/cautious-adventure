5.1.2. 3D (BCC) -> $\frac{N}{V} = \frac{2}{a^3}$

Using our Debye frequency as 
$$
\omega_D^3 = 6\pi^2 \frac{N}{V} v^3 = \frac{12\pi^2 v_s^3}{a^3}\\
\omega_D = \sqrt[3]{\frac{12\pi^2 v_s^3}{a^3}} = \frac{v_s}{a}\sqrt[3]{12\pi^2 } 
$$
we have 
$$
\omega_D = \frac{6e5}{3.5e-8}\sqrt[3]{12\pi^2} = 8.42\times 10^{13} \text{Hz} \\
=84.2 \text{Thz} \\
$$
For temperature, we'll use $k_B T_D = \hbar \omega $
$$
T_D = \frac{\hbar \omega}{k_B}  = 643 K
$$

5.2 We need to find $C=\frac{dE}{dT}$, so we need to find $\langle E \rangle(T)$

$$
\langle E \rangle = \int_0^\infty d \omega g(\omega) \hbar \omega n_B(\beta \hbar \omega)
$$
The only tough part is $g(\omega)$. In class we used $g(\omega) = N \frac{9\omega^2}{\omega_D^3}$, but that was in a 3D case.

Note that $N(\omega) = \int_0^{\omega_D} g(\omega) d\omega$, which comes out to 3N using the above formula for $g(\omega)$. The 3*N comes from the 3 polarization of phonons

In 1D we should have our $g(\omega)$ be independent of frequencey and only have N modes. I came up with

$$
g(\omega) = \frac{N}{\omega_D}
$$

Where $\omega_D = v\frac{\pi}{a}$ (which I stole from your formula for $T_D$)


Now we can do the integral!

$$
x = \beta \hbar \omega \\
dx = \beta \hbar d\omega \rightarrow d\omega = \frac{dx}{\beta \hbar}\\

\langle E \rangle = \int_0^\infty d\omega \frac{N}{\omega_D} \hbar \omega \frac{1}{e^{\beta\hbar \omega}-1} \\
= \int_0^\infty \frac{dx}{\beta \hbar} \frac{N}{\omega_D} \frac{x}{\beta} \frac{1}{e^{x}-1} \\
= \frac{1}{\beta^2 \hbar}\frac{N}{\omega_D} \int_0^\infty dx  \frac{x}{e^{x}-1} = \frac{k_B^2T^2}{ \hbar}\frac{N}{\omega_D} \frac{\pi^2}{6}
$$

Which gives us our heat capacity

$$
C = \frac{k_B^2T}{ \hbar}\frac{N}{\omega_D} \frac{\pi^2}{3} \\
= \frac{\pi^2k_B^2}{3\hbar} \frac{TNa}{v \pi} = \frac{\pi k_B^2}{3\hbar} \frac{Na}{v}T
$$

This has $N$ dependence, but I'm sure that isn't an issue. Also $Na=L$
