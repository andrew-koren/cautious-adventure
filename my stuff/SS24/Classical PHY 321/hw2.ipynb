6c) $\vec r(t) = x(t)\hat x + y(t)\hat y +z(t)\hat z$, with the magnitude being the root of the sum of each value squared.

$t_f = \dfrac{2v_0sin(\theta)}{g}$

I guess I can define a function with initial velocity and angle as inputs? Considering that there is no y-component, I'll produce a 2D graph of motion instead of a 3D graph of motion

```python
def golfball(theta,v0):
    r = []
    g = 9.81

    tf = 2*v0*np.sin(theta)/g  # The final time to be evaluated
    dt = 0.1  # The time step size
    t = np.arange(0,tf+dt,dt) # The time array
    
    r.append(v0 * np.cos(theta)*t)
    r.append(0)
    r.append(-g*t**2/2+v0 * np.sin(theta)*t)
    
    plt.scatter(r[0],r[2])
    plt.title(f'Golfball shot at {v0} m/s at angle {theta}')
    plt.xlabel('x')
    plt.ylabel('z')
    plt.show()
    
    return(r)
```

```python
theta = np.pi/180 * 30
v0 = 90

r = golfball(theta,v0)


print('the maximum values of x, y, and z are', r[0].max(),r[1],r[2].max(), 'respectively' )
```
