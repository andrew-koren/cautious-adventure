### Exercise 1 (10pt) Paths along a curved surface

We have shown by minimizing the path length on a 2D flat surface that we have generated a linear equation, which illustrates the shortest distance between two points is a straight line. On a curved surface, the shortest distance will be a curve of some kind. In the exercise you will be asked to show that the shortest distance between two points on a sphere and a cylinder, both of fixed radii.

A step in spherical coordinates is given by:

$$d\vec{s} = dr \hat{r} + r d\theta \hat{\theta} + r \sin\theta d\phi \hat{\phi}.$$

A step in cylindrical coordinates is given by:

$$d\vec{s} = d\rho \hat{\rho} + \rho d\phi \hat{\phi} + dz \hat{z}.$$

1. (5 pt) Demonstrate the integral for the shortest distance between two points on a sphere of radius $R$ is given by:

$$L = R \int_{\theta_1}^{\theta_2} \sqrt{1 + \left(\sin^2\theta\right) \left(\frac{d\phi}{d\theta}\right)^2} d\theta$$

where $\theta$ is the polar angle and $\phi$ is the azimuthal angle. Note that you want to travel from ($\theta_1, \phi_1$) to ($\theta_2, \phi_2$) at fixed $R$.

2. (5 pt) Construct the integral for the shortest distance between two points on a cylinder of radius $R$ and height $h$. The coordinates are $(R, \phi_1, z_1)$ and $(R, \phi_2, z_2)$ in the cylindrical coordinate system, where $\rho$ is the radial difference, $\phi$ is the azimuthal angle, and $z$ is the height.

Note the shortest path in cartesian is just $\int_a^b ds = \int_a^b\sqrt{dx^2 + dy^2}$

1.1. This has two variables, $\phi$ and $\theta$ since $r = R$ and $dr = 0$. To minimize L, we must minimize $L=\int_a^b ds$ where $d\vec s = R d\theta \hat \theta +R \sin \theta d\phi \hat \phi$

As such, the magnitude $ds = |d\vec s|$ is $$ ds = \sqrt{R^2 d\theta ^2 + R^2 \left(\sin \theta \right)^2 d\phi^2} = R \sqrt{ \frac{d\theta ^2}{d\theta^2} + \left(\sin \theta \right)^2 \frac{d\phi^2}{d\theta^2}} = R \sqrt{1 + \left(\sin^2\theta\right) \left(\frac{d\phi}{d\theta}\right)^2} d\theta$$

This allows us to find L using only an integral of $\theta$, so we can substitute this into the path-minimization equation

$$

As you can guess from the equations of motion, the center of mass is unaffected by the spring and drifts along, while the two masses oscillate in position relative to eachother around the equilibrium distance $L=4$

```python
import numpy as np
import matplotlib.pyplot as plt

x_cm0 = 0
x0 = 1
dx_cm0 = 2
dx0 = 3
L = 4
m = 1
k = 1

time = 10
steps = 1000
dt = time/steps
center_and_distance = np.zeros((steps,4))

def cmf(x,v,dt):
    dv = 0
    dx = v
    return x + dx*dt, v + dv*dt

def shof(x,v,dt):
    dv = - k * (x + L)
    dx = v
    return x + dx*dt, v + dv*dt

center_and_distance[0,:] = [x_cm0,dx_cm0,x0,dx0]

for i, vals in enumerate(center_and_distance):
    if i < steps-1:
        center_and_distance[i+1,:2] = cmf(vals[0],vals[1], dt)
        center_and_distance[i+1,2:] = shof(vals[2],vals[3], dt)
```

```python
fig, axs = plt.subplots(2,1,figsize=(8,10))
axs[0].set_title('Center of Mass')
axs[0].plot(range(len(center_and_distance)),center_and_distance[:,0], label = 'position')
axs[0].plot(range(len(center_and_distance)),center_and_distance[:,1], label = 'velocity')
axs[0].legend()
axs[1].set_title('Relative separation')
axs[1].plot(range(len(center_and_distance)),center_and_distance[:,2], label = 'position')
axs[1].plot(range(len(center_and_distance)),center_and_distance[:,3], label = 'velocity');
axs[1].legend();
```
