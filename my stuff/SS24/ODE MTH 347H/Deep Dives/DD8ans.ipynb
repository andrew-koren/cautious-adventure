# Deep Dive #8 - Integrating Factor Method

Remember, matricies aren't actually scary. They just say that the rate of change of a variable is dependent on the current value of other variables. 

The integrating factor method introduced in section 1.4 for constant coefficients is $y'=ay \implies y= ce^{at}$ which, with intial value $y_0=y(0)$ this becomes $y=y_0e^{at}$

How do we do this when A is a matrix? We'll use the property of matrix exponentials 
<mark> 
$\left(e^{At}\right)'=e^{At}A $ 
</mark> as follows

$$
x'=Ax \\
x'-Ax=\bold{0} \\
e^{-At}x'-e^{-At}Ax=e^{-At}\bold{0}\\
e^{-At}x'+e^{-At}(-A)x=\bold{0}\\
e^{-At}x'+\left(e^{-At}\right)'x=\bold{0}\\
\left(e^{-At}x\right)'=\bold{0} \\
e^{-At}x=\bold{c} \\
x(t)=e^{At}c \\
x(0)=e^0c=x^0\implies c=x^0\\
x(t)=e^{At}x^0
$$

$$
e^{At}x^0=e^{At}\left(c_1v_1+\cdot+c_nv_n\right)\\=e^{At}c_1v_1+\cdots+e^{At}c_nv_n=c_1e^{At}v_1+\cdots+c_ne^{At}v_n
$$

In deep dive #7 Question 6 we showed that $e^{A}v=e^{\lambda}v$. I'll copy-and-paste the proof in here, now with the addition of $t$.

Each $\lambda$ and $v$ correspond to a pair of $\lambda_i$ and $v_i$.
$$
e^{At} v = Iv + Atv + \frac{A^2t^2v}{2}+\dots+\frac{A^nt^nv}{n!} \\
=Iv + Avt + \frac{A^2vt^2}{2}+\dots+\frac{A^nvt^n}{n!} = Ivt + \lambda vt +  \frac{\lambda^2vt^2}{2}+\dots+\frac{\lambda^nvt^n}{n!}\\
= Itv + \lambda tv +  \frac{\lambda^2t^2v}{2}+\dots+\frac{\lambda^nt^nv}{n!} \\ =e^{\lambda t} v
$$
