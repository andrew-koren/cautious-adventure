(a)

$$
Z = \sum_{s} e^{-\epsilon_s/\tau} \
$$

This problem is just a fancy way of saying there are N+1 states with energy 0 to $\epsilon \times N$

$$
Z = 1 + e^{-\epsilon/\tau} + e^{-2\epsilon/\tau} +\dots + e^{-N \epsilon /\tau} \
\hspace{1.2cm} -Ze^{-\epsilon/\tau} = -e^{-\epsilon/\tau} - e^{-2\epsilon/\tau} - \dots - e^{-N \epsilon /\tau} - e^{-(N+1)\epsilon/\tau} \
Z - Ze^{-\epsilon/\tau} = 1 - e^{-(N+1)\epsilon/\tau} \
Z \left(1-e^{-\epsilon/\tau}\right) = 1 - e^{-(N+1)\epsilon/\tau} \
Z = \frac{1 - e^{-(N+1)\epsilon/\tau}}{1-e^{-\epsilon/\tau}}
$$

Yes I did align #1 and #2 on purpose

(b) $\epsilon \gg \tau \rightarrow \epsilon \beta \gg 1$

$$
Z = \frac{1 - e^{-(N+1)\epsilon\beta}}{1-e^{-\epsilon \beta}} \
\ln Z = \ln\left(1 - e^{-(N+1)\epsilon\beta}\right) - \ln\left(1-e^{-\epsilon \beta}\right) \
U = - \frac{\partial \ln Z}{\partial \beta} \
$$

```python
import sympy as sym
from sympy import ln, exp

epsilon, beta, N = sym.symbols('epsilon beta N')

lnZ = ln( ( 1 - exp(-(N+1)*epsilon*beta) ) / (1 - exp(-epsilon*beta) ) )
sym.diff(-lnZ, beta)

```

$$
- \frac{\left(1 - e^{- \beta \epsilon}\right) \left(- \frac{\epsilon \left(- N - 1\right) e^{\beta \epsilon \left(- N - 1\right)}}{1 - e^{- \beta \epsilon}} - \frac{\epsilon \left(1 - e^{\beta \epsilon \left(- N - 1\right)}\right) e^{- \beta \epsilon}}{\left(1 - e^{- \beta \epsilon}\right)^{2}}\right)}{1 - e^{\beta \epsilon \left(- N - 1\right)}} = 

- \frac{\left(- {\epsilon \left(- N - 1\right) e^{\beta \epsilon \left(- N - 1\right)}} 
- \frac{\epsilon \left(1 - e^{\beta \epsilon \left(- N - 1\right)}\right) e^{- \beta \epsilon}}{1 - e^{- \beta \epsilon}}\right)}{1 - e^{\beta \epsilon \left(- N - 1\right)}} \\

= -\frac{\epsilon (N+1)e^{-\epsilon \beta(N+1)}}{1-e^{-\epsilon \beta (N+1)}} - \frac{\epsilon e^{-\epsilon \beta}}{1-e^{-\epsilon \beta}}
$$

Note that we want the average number $n$ of open links, which has energy $\epsilon n$

$$
n = \frac{U}{\epsilon} = -\frac{(N+1)e^{-\epsilon \beta(N+1)}}{1-e^{-\epsilon \beta (N+1)}} - \frac{ e^{-\epsilon \beta}}{1-e^{-\epsilon \beta}} \\
= \frac{N+1}{1-e^{\epsilon \beta (N+1)}} + \frac{1}{1-e^{\epsilon \beta}}
$$

Aparently $n$ is negative? This is an exponential decay, and for N >>1 it basically converges to $(1-e^x)^{-1}$.
