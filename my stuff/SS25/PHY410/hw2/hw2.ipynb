(a) $\sigma(U) =  \ln(CU^{3N/2}) = \ln(C) + \frac{3N}{2}\ln(U)$

$$
\frac{1}{\tau} = \frac{\partial \sigma}{\partial U} = \frac{d}{dU} \ln\left(CU^{3N/2}\right) \\
\frac{1}{\tau} = \frac{3N}{2U} \\
U = \frac{3N}{2}\tau
$$

(b) Just take the derivative of $\frac{3N}{2U}$

$$
\frac{\partial}{\partial U}\frac{3N}{2U} = -\frac{3N}{2U^2}
$$

Since $N$ is positive, this must be negative

Note $B/2m = \langle s \rangle$, and if necessary $g(N,0)\approx \sqrt{\frac{2}{\pi N}} 2^N$

$$
\frac{2s^2}{N} = \frac{U^2}{2m^2} B^2 N \\
s^2 = \frac{U^2}{2^2m^2} B^2 N^2 \\
s = \frac{U}{2m} BN
$$
