Mean Field Ferromagnet: 2nd order phase transition

Mean field $B_E = \lambda M$ <br>
Replace $M$ with $\lambda M$

$$
M = n \mu + \tanh \left( \frac{\lambda \mu B}{\tau}\right) 
$$
Dimensionless:
$$
m \equiv \frac{M}{n\mu} \\
t = \frac{\tau}{n\mu^2 \lambda} \\
m = \tanh \left(\frac{m}{t} \right)
$$

From the solution below, we can have either paramagnetic or diamagnetic region. Transition at $t_c = 1\rightarrow \tau_c = n \mu^2 \lambda$. Generally there is a ordered and disordered state, where the interactions in the material will cause it to be ordered.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1.2, 1000)
m = np.linspace(0, 1.2, 1000)

plt.plot(x, m)

for t in [0.01, .5, 0.9, 1, 2]:
    m2 = np.tanh(m/t)
    plt.plot(x, m2)

plt.xlabel('arb')
plt.ylabel(r'$m\equiv \frac{M}{n\mu}$');
```
