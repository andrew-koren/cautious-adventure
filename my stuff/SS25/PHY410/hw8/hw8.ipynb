$$

\mu(R) = \tau \ln \left( \frac{n(R)}{n_Q}\right) - MgR

$$
\mu(r) = \mu(R) \\
\tau \ln\left( \frac{n(r)}{n_Q}\right) - \frac{MgR^2}{r} = \tau \ln \left( \frac{n(R)}{n_Q}\right) - MgR \\
\ln\left( \frac{n(r)}{n_Q}\right)-\ln \left( \frac{n(R)}{n_Q}\right) = \frac{MgR^2}{r\tau}- \frac{MgR}{\tau} \\
\frac{n(r)}{n(R)} = e^{\frac{MgR^2}{r\tau}- \frac{MgR}{\tau}}
