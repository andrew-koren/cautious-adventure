(a) 1D electron energy 

$$
\epsilon_n = \frac{\hbar^2}{2M} \left(\frac{\pi n}{L}\right)^2
$$

Where we choose $n$. I believe $n$ is actually the number of particles in the box too (x2 since two per orbital)


$$
N(\epsilon) = \frac{2L}{\pi} \sqrt{\epsilon \frac{2M}{\hbar^2} } \\
= \frac{2L}{\hbar \pi} \sqrt{2M\epsilon}
$$
This gives us our density of states
$$
\mathcal D(\epsilon) \equiv \frac{dN}{d\epsilon} = \frac{2L}{\hbar \pi} \sqrt{2M\epsilon} \frac{1}{2 \epsilon} \\
= \frac{L}{\hbar \pi} \sqrt{\frac{2M}{\epsilon}}
$$

(b) ground state energy. Depends on fermi energy

$$
U_0 = \int_0^{\epsilon_F} \epsilon \mathcal D(\epsilon) d\epsilon \\
= \frac{L}{\hbar \pi} \sqrt{2M} \int_0^{\epsilon_F} \sqrt{\epsilon} d\epsilon
= \frac{L}{\hbar \pi} \sqrt{2M} \frac{2 \epsilon_F^{3/2}}{3 } \\
= \frac{2\sqrt{2}}{3}\frac{L}{3\hbar \pi} \sqrt{M\epsilon_F^3}
$$

This can be converted back to $N$ using the above equation for $\epsilon_n$ and $N=2n$

(c) 2D. Not so bad, just need to redefine $n$ and $N$.

A 1/4 circle with $n$-radius has

$$
N = 2 \times 1/4 \times \pi n^2 \\
n = \sqrt{\frac{2N}{\pi}} \\
\epsilon_n = \frac{\hbar^2}{2M} \left(\frac{\pi n}{L} \right)^2 = \frac{N \hbar^2 \pi}{ML^2}  \\
N = \frac{ML^2\epsilon}{\pi \hbar^2} \\
\frac{\partial N}{\partial \epsilon} = \frac{ML^2}{\pi \hbar^2}
$$

(d) Use the same procedure as before to find our ground state

$$
U_0 = \int_0^{\epsilon_F} \epsilon \mathcal D(\epsilon) d\epsilon \\
= \frac{ML^2 \epsilon_F^2}{2\pi \hbar^2}
$$

```python
import sympy as sp
from sympy.abc import epsilon, n, N, L, M
hbar, eF = sp.symbols('hbar epsilon_F')

n = sp.sqrt(2*N/sp.pi)
en = hbar**2/(2*M) * (sp.pi*n/L)**2
DOS = sp.solve(sp.Eq(epsilon, en), N)[0]/epsilon

U0 = sp.integrate(epsilon*DOS, (epsilon, 0, eF))
U0
```
