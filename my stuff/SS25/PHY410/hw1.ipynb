Using binomial $\delta n = \sqrt{np(1-p)}$, where $p=0.5$, we have
$$
\begin{matrix}
N=12 & \delta n = \sqrt{12*.25} = \sqrt{3} \approx 1.7
\\ & \frac{\delta n}{\langle n \rangle} = \sqrt{3}/6 \approx 0.29
\\
N=400 & \delta n = \sqrt{400*.25} = 10
\\ & \frac{\delta n}{\langle n \rangle} = 10/200 = 0.05
\end{matrix}
$$
From this we can tell it's much less likey to obtain fewer than 30% for N=400 compared to N=12, as the relative uncertainty is much smaller. 

Using a binomial distribution calculator, we see the odds for N=12 <30% (0-3 coins) is .07, but for N=400 <30% (0-119 coins) is $10^{-16}$
