**Introduction**

In this experiment, we aim to explore the principles of spectroscopy by analyzing the spectra of atomic gases. The primary objective is to use two different dispersion mechanisms—a prism and a diffraction grating—to separate light into its component wavelengths. By doing so, we can make accurate measurements using an optical spectrometer, calibrate our instruments, and investigate the relationship between wavelength, refraction, and diffraction.

Spectroscopy is a fundamental tool in atomic physics, as it allows us to probe the internal structure of matter by studying the emission and absorption spectra of different elements. The discrete spectral lines observed in this experiment provide insight into the quantized energy levels of atoms, reinforcing the principles of quantum mechanics. Using the spectrometer, we will measure spectral line positions, calculate wavelengths, and compare our findings with theoretical predictions, such as those derived from the Rydberg formula.

**5. The Prism Spectrometer**

1. **Introduction**

   The purpose of this section is to analyze the dispersion of light through a prism spectrometer and determine the index of refraction for different wavelengths of helium. The independent variable in this experiment is the wavelength of light, while the dependent variable is the deviation angle, $\beta$. The index of refraction $n$ is calculated using the following equation:

   $n(\beta) = \frac{\sin \left(\frac{\alpha + \beta}{2} \right)}{\sin \left(\frac{\alpha}{2} \right)}$

   The prism will be calibrated using known helium spectral lines, allowing us to determine the constants $A$ and $B$ in Cauchy’s equation:

   $n(\lambda) = A + \frac{B}{\lambda^2}$

13. **Application to Solar Spectroscopy**

   The prism spectrometer can be used to analyze the solar spectrum by measuring the absorption lines present in sunlight. These absorption lines, known as Fraunhofer lines, correspond to elements present in the Sun’s outer layers. By comparing the measured absorption lines with known atomic spectra, it is possible to determine the elemental composition of the Sun's atmosphere.

**6. The Grating Spectrometer**

1. **Introduction**

   A diffraction grating disperses light by means of interference, creating bright, widely separated maxima corresponding to specific wavelengths. The diffraction equation that governs this behavior is:

   $d \sin \theta = m \lambda$

   where $d$ is the grating spacing, $\lambda$ is the wavelength, $m$ is the order number, and $\theta_m$ is the diffraction angle. As such, $\theta$ is the independent and $\lambda$ is the dependent variable.

2. **Justification for Using First-Order Maxima**

   The first-order maxima ($m = 1$) are used because they are the brightest. Higher-order maxima are almost impossible to see let alone measure with our lab setup.

3. **No Calibration Required**

   Unlike the prism spectrometer, a diffraction grating does not require calibration with known spectral lines. The diffraction angles directly correspond to the wavelengths of light according to the grating equation, making the measurement process more straightforward.

4. **Experimental Procedure**

   The following describes the procedure for alignment:

   1. Focus on the helium tube. 
   Place the lamp as near to the collimator as is safe and open the collimator until light is visible. Reduce the collimator and refocus by adjusting the eyepiece coarse and fine focusing knobs. Continue until the light is a thin yellow sliver

   2. Align the diffraction grating. 
   With the base and telescope unlocked and set near zero, place the diffraction grating down. Adjust the base-telescope alignment until the central yellow line is at zero degrees. For precise tuning, make sure the left and right side of the yellow line correspond to the same arc minute left and right on the venier scale. Lock the base.

**Summary**

   See above tables for data on each gas. Fractional uncertainties and percent error is also included.

- The prism spectrometer results showed small deviations from theoretical values, with relative errors below 4%.
- The diffraction grating results were more precise, with percent errors generally below 3%.
- The grating method proved to be more accurate and easier to use due to its lack of dependence on calibration.
- Crosshairs were often difficult to observe, especially with dimmer colors (blue, violet). Further gains in accuracy are possible if low-light cameras or other methods are used to detect 2nd order fringes and take more accurate measurements.
