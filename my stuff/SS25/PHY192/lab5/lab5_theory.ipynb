(3)

a. $\lambda = \frac{1}{\text{time}}$

b. See below. The first curve $2\lambda$ decays much faster

```python
from numpy import linspace, exp, log
import matplotlib.pyplot as plt

t = linspace(0, 5, 1000)
lambda_list = [2, 1, 1/2]

for lbd in lambda_list:
    plt.plot(t, exp(-lbd*t), label=fr'{lbd}$\lambda_0$')

plt.xlabel('time (arb)')
plt.grid()
plt.legend();
```

c. you can linearize by taking the negative log of the right side, then the slope is the decay constant $\lambda$

```python
for lbd in lambda_list:
    plt.plot(t, -log(exp(-lbd*t)), label=fr'{lbd}$\lambda_0$')

plt.xlabel('time (arb)')
plt.grid()
plt.legend();
```

In terms of quantitites, this means 

$$
N = e^{-\lambda t} \\
-\ln(N) = \lambda t
$$
