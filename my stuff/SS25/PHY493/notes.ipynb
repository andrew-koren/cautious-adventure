### Quarks

6 types - upper and lower <br>
all have: mass, 1/2 spin

up $u$, charm $c$, top $t$ +2/3 charge <br>
down $d$, strange $s$, bottom $b$ -1/3 charge

### Leptons

6 types - normal or neutrino <br>
All have: mass, 1/2 spin


electron $e$, muon $\mu$, tau $\tau$ carry their flavors and -1 charge <br>
neutrinos $\nu$ have flavors but no charge
