Fermi's golden rule

$$
W = \frac{2\pi}{\hbar} | \mathcal M_{if} |^2 \rho(E) \\
W \text{ transition rate} \\
\mathcal M_{if} \text{ dynamic info} \\
\rho(E) \text{ Phase Space / Density of States}
$$

Density of states describes states which can occur.

If a particle can have momentum $p = \hbar k = h / \lambda$, then its momentum space is $N_i = \frac{V}{(2\pi \hbar)^3} \int d^3 p$. Density comes from the fact that volume is here, which comes from $V = \int dx dy dz$ (arb. bounds).

This is just generic math so far (also position/momentum aren't related yet). Physics: Momentum conservation.

Actual (fewer) states:
$$
\int d^3 p \rightarrow \int d^3 p \delta(p_i - \sum p_f) \\
\text{Only evaluates where final and inital momentum are equivalent}
$$

Okay, but we need some more physics. 4-momentum has lorentz invariance

$$
N_n = \frac{1}{(2\pi\hbar)^{4n}} \int \Pi_{i=1}^n d^4 p_i \delta \left( \left( p^\mu p_\mu \right)_i - m_i^2 c^2 \right) \\
\text{Garuntees on-shell (real, not virtual)}
$$

Moar physics. Energy/momentum conservation (partially implemented above), also step function on locations where energy < 0.

The above equations are for unit space V=1, for general wave space instead of momentum space we fourier transform.

$$
\rho(E) = \frac{\partial N_n}{P\partial E} = \frac{(2\pi)^{n+4}}{(2\pi \hbar)^{4n}} \delta^4 \left( p_i - \sum p_f \right) \Pi_{j=1}^n \dots
$$

For center of momentum frame with two particles $\rho(E) = \frac{\pi}{(2\pi)^6}\frac{p_1}{E}$

Klein-Gordon with new fancy notation

$$

- \frac{\partial \psi}{\partial t^2} + \nabla^2 \psi = m^2 \psi \\

$$

4-vector to describe solutions!
$$
\psi(x,t) = e^{i p_\mu x^\mu} = e^{-i Et + ip\cdot x} \\
p_\mu x^\mu = \hbar (k\cdot x - \omega t) \\
\text{KG} \Rightarrow -E^2 + p^2 +m^2 = 0 \Rightarrow E = \pm \sqrt{p^2 + m^2}
$$
BTW negative energy indicates antiparticles. Don't worry about it.

A more important issue: probabilities. Continuity is

$$
\frac{\partial}{\partial t} \rho + \nabla \cdot j = 0 \rightarrow \text{ works for Schrodinger} \\

$$
where $\rho$ is probability density.

Maybe our problem is $\square$? Let's see what Dirac found.

$$
\left( p^\mu p_\mu - m^2 c^2 \right) \dots \text{ invariant} \\
