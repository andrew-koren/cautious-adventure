$$

Comes out to $1.85\times10^9$ alpha particles per second

    ii. How many electrons-ion pairs are produced per alpha?
Bringing a $5$-MeV particle to stopped (ignoring 3.7GeV rest energy for $\alpha$-particles) by creating $30$-eV particles would create
$$
5\times 10^6 /30 = 1.67 \times 10^5 e/\alpha \\
(1.67 \times 10^5)(1.85\times10^9) = 3.09 \times 10^{14} e/s
$$
Which also checks off iii.

= 2.75 \times 1.602 \times 10^{-5} C/s = 4.406 \times 10^{-5} A

$$

(b) What is the velocity of $^{16}O$ at $80\text{ MeV}$? 

Low compared to GeV rest energy, nonrelativistic. Start with $c=1$

$$
v = \frac{p}{m} = \frac{80 \text{MeV}}{16u*931.5\text{MeV/u}} = \frac{80}{14927.7} = 0.00537 \\
c = 3e8 \rightarrow v = 1.6\times 10^6 m/s \\
\Delta x = 1.6\times 10^6\times 1.65\times 10^{-8} = 2.66\text{cm}
$$
