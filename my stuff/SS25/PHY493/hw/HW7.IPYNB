$v$ is easy to calculate, can we plug in $E$ and have $m$ fall out?

Oh wait, there's multiple velocities and multiple energies. Not FTL.

$$
L = v*t \\
\Delta t = \frac{L}{v_{\text{slow}}} - \frac{L}{v_{\text{fast}}} \\
v_{\text{slow}} = v_1 \\
v_{\text{fast}} = v_2 \\

\frac{\Delta t}{L} = \frac{1}{c} \left[1 + \frac{1}{2}\left(\frac{mc^2}{E_1} \right)^2 \right] - \frac{1}{c}\left[1 + \frac{1}{2}\left(\frac{mc^2}{E_2} \right)^2 \right] = \frac{(mc^2)^2}{2c} \left( \frac{1}{E_1^2} - \frac{1}{E_2^2} \right) \\
\left(mc^2\right)^2 = \frac{2c \frac{\Delta t}{L}}{\left( \frac{1}{E_1^2} - \frac{1}{E_2^2} \right)}
$$

Don't forget to convert light years to... light seconds? We're in natural units so we want to stay away from meters.

$$
m_0c^2 = 52\times 10^{-6} \text{MeV} = 52\text{eV}
$$

```python
c = 1
dt = 10/(365*24*60*60)
L  = 1.7e5
E1 = 20
E2 = 30
mc22 = (2*c*dt/L)/((1/(E1**2))-(1/(E2**2)))
mc2 = mc22**(1/2)
mc2
```

The hint equation comes from (9.34) in the textbook, and we're using $m_\mu \ll m_\tau$

$$
\tau_\tau = \left(\frac{M_W}{m_\tau g_w} \right)^4 \frac{12 \hbar (8\pi)^3}{m_\tau c^2} \\
g_w = 0.653 \\
M_W = 80.379 \text{ GeV}/c^2
$$

Since the $\tau$ can decay into both muons and electrons, it's lifetime is halved ($\Gamma$ is doubled, electron and muon mass are similar $\ll m_\tau$ so they have equal probability)

```python
from scipy.constants import pi

M_W = 80.379
g_w = 0.653
m_tau = 1.777
c=1
hbar = 1

tau_tau = (M_W/(m_tau*g_w))**4*12*hbar*(8*pi)**3/(m_tau*c**2)/2
tau_tau
```

No need to panic! This isn't in units of seconds, it is $\frac{1}{\text{GeV}}$

```python
tau_tau*(1.52 * 10**24)
```

Okay maybe now panic. Let's try again in SI. The units in the $()^4$ term cancel, but $\frac{1}{m_\tau}$ should be converted
$$
m_\tau (\text{GeV}/c^2) \rightarrow m_\tau (\text{J})
$$

```python
from scipy.constants import hbar, c
GeV_to_Joule = 1.60218e-10

m_tau_si = m_tau *GeV_to_Joule/c**2

tau_tau = (M_W/(m_tau*g_w))**4*12*hbar*(8*pi)**3/(m_tau_si*c**2)/2
tau_tau
```

This is ~3x the value in the PDG, which is likely due to all the non-leptonic ways that $\tau$ can decay, which makes sense since it is as massive as some nuclei.
