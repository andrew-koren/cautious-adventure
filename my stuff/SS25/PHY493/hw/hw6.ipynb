The muons and $\Mu$ notation are kinda clashing, so I changed muons to be $\Mu$. So far, no features of muons or electrons have been used, so the ame matrix element can be used for this cross section

$$
M = -\frac{g_e^2 Q_q}{(p_1+p_2)^2} \left[ \bar v_e \gamma^\mu u_e \right]
\left[ \bar u_p \gamma_\mu v_p \right] = 
-\frac{g_e^2 Q_q}{s} \left[ \bar v_e \gamma^\mu u_e \right]
\left[ \bar u_p \gamma_\mu v_p \right]
$$

Notice that the factor $Q_q$ has been added by the quark vertex, which can be $-1/3$ or $+2/3$ depending on the quark flavor.

In Hw5 we used particle properties to apply Casimir's trick

- $E_e \gg m_\mu$
- sum over e, muon spins

Each quark flavor only has two spin states, and we have the same high energy condition, so this result is also retained

$$
\langle | M | ^2 \rangle = \frac{4 g_e^4 Q_q^2}{s^2} \left(E^2(1-\cos \theta) \right)^2 + \left(E^2 (1+\cos \theta)\right)^2 \\ 
= g_e^4 Q_q^2 \left( 1 + \cos^2 \theta \right)
$$

From this we can find our scattering cross section

$$
\frac{d\sigma}{d\Omega} = \frac{1}{(8\pi)^2} \frac{\langle | M | ^2 \rangle}{(E_1 + E_2)^2} \frac{|p_f |}{|p_i|} \\
= 
\frac{g_e^4 Q_q^2}{256 \pi^2 E^2} \left(1 + \cos^2 \theta \right) \\
\sigma = \frac{g_e^4 Q_q^2}{48 \pi E^2}
$$

(c) We can combine all six quark cross sections to compute the total cross section. With our high energy assumption, the masses of each quark doesn't affect the outcome, only charge.
