$Z$ same as $Z^0$?

answers

(a)
- energy conservation <br>
The photon cannot create mass by itself
- lepton number <br>
lepton number is violated, $L_e$ and $L_\mu$ go from (0,0) to (1,-1)

(b) 
- energy conservation <br>
The $W^+$ boson doesn't have enough mass for the top quark $80.39 GeV/c^2 < 173.1 GeV/c^2$

Looks like everything else is conserved, so some additional energy could make this possible (off-shell)

(c) 
- energy conservation (acceptable) <br>
The $Z^0$ boson mass is much larger than the rest mass of each muon $91.19 >> 0.212$. This is fine as long as excess energy becomes kinetic
- lepton number <br>
would be fine if $\mu^+$ and $\mu^-$, instead we have +2 $L_\mu$ after the decay
- charge conservation <br>
$0 \rightarrow +2$

(d)
- energy conservation (acceptable) <br>
Excess energy must go somewhere. The bottom quark has 4.18 GeV/c^2, and $173.1 > 80.39 + 4.18$

Flipping the antibottom quark to bottom quark means that everything else is conserved, same as in (b)

(e)
- energy conservation (acceptable) <br>
$W^-$ is much more massive than the electron and neutrino, so excess mass must be converted to kinetic energy $80.39 > 0.000511 + \nu_e$
- lepton number <br>
$L_e$ goes from zero to two, since a regular neutrino is produced. An antineutrino would balance this.

(f) 
- energy conservation <br>
On-shell photons cannot produce mass. More like momentum conservation.
