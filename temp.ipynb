## Polarized sphere

1. 
$$
\vec P = P_0 \vec r = P_0 r \hat r \\
\sigma_b = \vec P \cdot \hat n = \left. P_0 r (\hat r \cdot \hat r)\right|_{r=a} =  P_0 a \\
\rho_b = - \nabla\cdot\vec P = -\frac{1}{r^2} \frac{d}{dr}P_0r^3 = -3P_0
$$
Both values are independent of our coordinates, so our integrals are quite easy

2. 
First $r<a$

$$
E*4\pi r^2 = \frac{-3P_0}{\epsilon_0} \int_V  dV =  \frac{-4\pi r^3}{\epsilon_0} \\
E(r) = -\frac{P_0}{\epsilon_0}\vec r
$$

Next: $r>a$.  We'll show there's no charge

$$
Q = \int_V \rho_b dV + \int_S \sigma_b da = -4P_0\pi a^3 + P_0 a*4\pi a^2 = 0\\
E = 0
$$

The electric field points entirely inward.
